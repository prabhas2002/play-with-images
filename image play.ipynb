```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# load the image
img_path = 'Koalainputimage.jpg'
img = Image.open(img_path)

# convert PIL image to numpy array
img_np = np.array(img)
```

```python
img_np.shape #rgb image
```

```python
img_np[100][10]
```

```python
plt.imshow(img_np)
```

```python
# plot the pixel values
plt.hist(img_np.ravel(), bins=50, density=True)
plt.xlabel("pixel values")
plt.ylabel("relative frequency")
plt.title("distribution of pixels")
```

```python
#grey scale
grey_img=np.zeros([768,1024])
#lighting 
li_img=np.zeros([768,1024])
# Luminosity Method
lu_img=np.zeros([768,1024])
```

```python
grey_img.shape
```

```python
#grey scale
for i in range(768):
    for j in range(1024):
        grey_img[i][j]=img_np[i][j].mean()
```

```python
#lightness method 
for i in range(768):
    for j in range(1024):
        li_img[i][j]=(min(img_np[i][j])+max(img_np[i][j]))/2
```

```python
#lightness method 
for i in range(768):
    for j in range(1024):
        lu_img[i][j]=0.3*img_np[i][j][0]+0.59*img_np[i][j][1]+0.11*img_np[i][j][2]
```

```python
# display the grey scale image
plt.imshow(grey_img)
```

```python
plt.imshow(li_img)
```

```python
plt.imshow(lu_img)
```

```python
#blur image
```

```python
grey_img.shape
```

```python
kernal=np.ones([5,5]) #to reduce resolution in image
big_kernal=np.ones([20,20])
out=np.zeros([764,1020]) #n-k+1 sizes
out1=np.zeros([749,1005])
```

```python
for i in range(764):
    for j in range(1020):
        for k in range(5):
              for m in range(5):
                    out[i,j]+=grey_img[i+k,j+m]*kernal[k,m]
```

```python
plt.imshow(out)
```

```python
for i in range(749):
    for j in range(1005):
        for k in range(20):
              for m in range(20):
                    out1[i,j]+=grey_img[i+k,j+m]*big_kernal[k,m]
```

```python
plt.imshow(out1) #more blurred image
```

```python
#edge detection
```

```python
edge_kernal=np.zeros([3,3])
edge_kernal[0,0]=0
edge_kernal[0,2]=0
edge_kernal[2,0]=0
edge_kernal[2,2]=0
edge_kernal[1,1]=-8
edge_kernal[0,1]=2
edge_kernal[1,0]=2
edge_kernal[1,2]=2
edge_kernal[2,1]=2
```

```python
out_edge=np.zeros([766,1022])
```

```python
for i in range(766):
    for j in range(1022):
        for k in range(3):
              for m in range(3):
                    out_edge[i,j]+=grey_img[i+k,j+m]*edge_kernal[k,m]
```

```python
plt.imshow(out_edge) #edges are being detected in this image
```

```python
ker=np.ones([3,3,3])
out_edge=np.zeros([766,1022])
```

```python
#blur original imge 
for i in range(766):
    for j in range(1022):
        for c in range(3):
            for k in range(3):
                  for m in range(3):
                        out_edge[i,j]+=img_np[i+k,j+m,c]*ker[k,m,c]
```

```python
plt.imshow(out_edge)
```
